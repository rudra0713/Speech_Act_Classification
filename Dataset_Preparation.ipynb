```python
import re

def prepare_txt():
    f = open('data/ArSAS-normalized.txt')
    all_text = open("data/dataset_full.txt", "w")
    speech_act_labels = ['Assertion', 'Recommendation', 'Expression', 'Question', 'Request', 'Miscellaneous']
    sentiment_labels = ['Positive', 'Negative', 'Neutral', 'Mixed']

    count = 0
    check_arr = []
    duplicate = 0

    line = f.readline()
    i = 0
    no_normalized = 0
    while line:
        try:
            line = f.readline()
        except:
            break
        tokens = re.split(r'\t+', line)
        try:
            tweet_id = int(tokens[0].replace('"', ''))
            tweet_30 = tokens[7].replace('"', '').split(" ")[:30]
            tweet_50 = tokens[7].replace('"', '').split(" ")[:50]
        except:
            no_normalized += 1
            continue
        label_sa = speech_act_labels.index(tokens[5].replace('"', ''))
        label_se =  sentiment_labels.index(tokens[3].replace('"', ''))


        if label_sa == 5:
            continue
        if label_se == 3:
            continue
        if tweet_id in check_arr:
            continue
        else:
            check_arr.append(tweet_id)
        all_text.write(line)
        i = i + 1
    all_text.close()
    print("total dataset length ", len(check_arr))
    return


prepare_txt()
```

```python
import csv, random, re

speech_act_labels = ['Assertion', 'Recommendation', 'Expression', 'Question', 'Request']
sentiment_labels = ['Positive', 'Negative', 'Neutral']

def arr_to_str(arr):
    return ' '.join(t for t in arr).lstrip('"').rstrip('"')


def all_text_field_to_file():
    global speech_act_labels
    f = open('data/dataset_full.txt')
    train_j = open("ArSAS/json/train_final.json", "w")
    test_j = open("ArSAS/json/test_final.json", "w")
    dev_j = open("ArSAS/json/dev_final.json", "w")
    
    train_sa_30 = csv.writer(open("ArSAS/data_sa_30/train.csv", "w+"))
    test_sa_30 = csv.writer(open("ArSAS/data_sa_30/test.csv", "w+"))
    test_sa_30_backup = csv.writer(open("ArSAS/data_sa_30/test_backup.csv", "w+"))
    dev_sa_30 = csv.writer(open("ArSAS/data_sa_30/dev.csv", "w+"))
    
    train_sa_50 = csv.writer(open("ArSAS/data_sa_50/train.csv", "w+"))
    test_sa_50 = csv.writer(open("ArSAS/data_sa_50/test.csv", "w+"))
    test_sa_50_backup = csv.writer(open("ArSAS/data_sa_50/test_backup.csv", "w+"))
    dev_sa_50 = csv.writer(open("ArSAS/data_sa_50/dev.csv", "w+"))


    train_se_30 = csv.writer(open("ArSAS/data_se_30/train.csv", "w+"))
    test_se_30 = csv.writer(open("ArSAS/data_se_30/test.csv", "w+"))
    test_se_30_backup = csv.writer(open("ArSAS/data_se_30/test_backup.csv", "w+"))
    dev_se_30 = csv.writer(open("ArSAS/data_se_30/dev.csv", "w+"))

    train_se_50 = csv.writer(open("ArSAS/data_se_50/train.csv", "w+"))
    test_se_50 = csv.writer(open("ArSAS/data_se_50/test.csv", "w+"))
    test_se_50_backup = csv.writer(open("ArSAS/data_se_50/test_backup.csv", "w+"))
    dev_se_50 = csv.writer(open("ArSAS/data_se_50/dev.csv", "w+"))
    
    test_sa_30.writerow(["id", "sentence"])
    test_sa_50.writerow(["id", "sentence"])
    test_se_30.writerow(["id", "sentence"])
    test_se_50.writerow(["id", "sentence"])

    test_sa_30_backup.writerow(["id", "sentence", "label"])
    test_sa_50_backup.writerow(["id", "sentence", "label"])
    test_se_30_backup.writerow(["id", "sentence", "label"])
    test_se_50_backup.writerow(["id", "sentence", "label"])



    train_count = 0
    dev_count = 0
    test_count = 0

    i = 0
    no_normalized = 0
    lines = f.readlines()
    for line in lines:
        tokens = re.split(r'\t+', line)
        tweet_id = int(tokens[0].replace('"', ''))
        tweet_30 = tokens[7].replace('"', '').split(" ")[:30]
        tweet_50 = tokens[7].replace('"', '').split(" ")[:50]
        label_sa = speech_act_labels.index(tokens[5].replace('"', '')),
        label_se = sentiment_labels.index(tokens[3].replace('"', ''))            

        ob = {
            'tweet_id': tweet_id,
            'tweet_30': tweet_30,
            'tweet_50': tweet_50,
            'label_sa': speech_act_labels.index(tokens[5].replace('"', '')),
            'label_se': sentiment_labels.index(tokens[3].replace('"', ''))            
        }

        rand = random.random()
        if rand > 0.4:
            train_j.write(str(ob).replace("\'", "\"") + '\n')
            train_sa_30.writerow([ob['tweet_id'], ob['label_sa'], "a", arr_to_str(ob['tweet_30'])])
            train_sa_50.writerow([ob['tweet_id'], ob['label_sa'], "a", arr_to_str(ob['tweet_50'])])
            train_se_30.writerow([ob['tweet_id'], ob['label_se'], "a", arr_to_str(ob['tweet_30'])])
            train_se_50.writerow([ob['tweet_id'], ob['label_se'], "a", arr_to_str(ob['tweet_50'])])
            train_count += 1
        else:
            if 0<rand<=0.2:
                dev_j.write(str(ob).replace("\'", "\"") + '\n')
                dev_sa_30.writerow([ob['tweet_id'], ob['label_sa'], "a", arr_to_str(ob['tweet_30'])])
                dev_sa_50.writerow([ob['tweet_id'], ob['label_sa'], "a", arr_to_str(ob['tweet_50'])])
                dev_se_30.writerow([ob['tweet_id'], ob['label_se'], "a", arr_to_str(ob['tweet_30'])])
                dev_se_50.writerow([ob['tweet_id'], ob['label_se'], "a", arr_to_str(ob['tweet_50'])])
                dev_count += 1
            else:
                test_j.write(str(ob).replace("\'", "\"") + '\n')
                test_sa_30.writerow([ob['tweet_id'], arr_to_str(ob['tweet_30'])])
                test_sa_50.writerow([ob['tweet_id'], arr_to_str(ob['tweet_50'])])
                test_se_30.writerow([ob['tweet_id'], arr_to_str(ob['tweet_30'])])
                test_se_50.writerow([ob['tweet_id'], arr_to_str(ob['tweet_50'])])
                
                test_sa_30_backup.writerow([ob['tweet_id'], arr_to_str(ob['tweet_30']), ob['label_sa']])
                test_sa_50_backup.writerow([ob['tweet_id'], arr_to_str(ob['tweet_50']), ob['label_sa']])
                test_se_30_backup.writerow([ob['tweet_id'], arr_to_str(ob['tweet_30']), ob['label_se']])
                test_se_50_backup.writerow([ob['tweet_id'], arr_to_str(ob['tweet_50']), ob['label_se']])
                test_count += 1
        i = i + 1
    train_j.close()
    test_j.close()
    dev_j.close()


    print("line count ", i)
    print("no normalized ", no_normalized)
    print("train ", train_count)
    print("dev ", dev_count)
    print("test ", test_count)

    
all_text_field_to_file()


```
