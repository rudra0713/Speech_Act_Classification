```python
from sklearn.metrics import precision_recall_fscore_support as score
import numpy as np, csv
import pandas as pd, re


speech_act_labels = ['Assertion', 'Recommendation', 'Expression', 'Question', 'Request', 'Miscellaneous']
sentiment_labels = ['Positive', 'Negative', 'Neutral', 'Mixed']


df_results = pd.read_csv("test_results.tsv",sep="\t",header=None)
df_results_csv = pd.DataFrame({'Is_Response':df_results.idxmax(axis=1)})

classification = []
for row in df_results_csv.itertuples():
    a,b = row
    classification.append(b)
print(len(classification))
print(classification[:5])

f = open('data/ArSAS-normalized.txt')
line = f.readline()

ground_truth = []
with open('data/test_30_sa.tsv') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    for i,row in enumerate(reader):
        if i == 0:
            continue
        id = int(row[0])
        while line:
            try:
                line = f.readline()
            except:
                break
            tokens = re.split(r'\t+', line)
            tweet_id = int(tokens[0].replace('"', ''))
            if id == tweet_id:
                ground_truth.append(speech_act_labels.index(tokens[5].replace('"', '')))
                break
print(ground_truth[:5])

precision, recall, fscore, support = score(np.array(ground_truth), classification)
print("Detailed evaluation:")
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))

```
